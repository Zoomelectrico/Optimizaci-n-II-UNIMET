### Pregunta 1

Considere la red de distribucin que muestra la Figura 1. Encuentre la ruta ms corta desde el nodo 1 hasta el nodo 8 de la red.
 1. Diga cual es la ruta y cuanto es la distancia recorrida.
 2. Considere que la ruta encontrada en **1** no es transitable, cual sera la modificacin que hara en el cdigo para buscar una nueva ruta ms corta.
 3. Cual es la nueva ruta ms en funcin de la condicin en **2**
 


**Figura 1. Red del Problema 1**

Fuente: Elaboracin Propia

````python
# Parte 1
G = nx.Graph()

G.add_edge(1, 2, distance=6)
G.add_edge(1, 3, distance=3)
G.add_edge(1, 4, distance=1)

G.add_edge(2, 3, distance=1)
G.add_edge(2, 5, distance=4)
G.add_edge(2, 6, distance=3)

G.add_edge(3, 4, distance=3)
G.add_edge(3, 5, distance=5)
G.add_edge(3, 6, distance=6)
G.add_edge(3, 7, distance=2)

G.add_edge(4, 6, distance=5)
G.add_edge(4, 7, distance=5)

G.add_edge(5, 6, distance=3)
G.add_edge(5, 8, distance=1)

G.add_edge(6, 7, distance=1)
G.add_edge(6, 8, distance=3)

G.add_edge(7, 8, distance=6)

path = nx.shortest_path(G, weight="distance", source=1, target=8)
print(path)
````

````python
# Parte 2
G.remove_edge(1, 4)
G.remove_edge(4, 6)
G.remove_edge(6, 8)
````

````python
# Parte 3
path = nx.shortest_path(G, weight="distance", source=1, target=8)
print(path)
````

Respuestas:

- Parte 1: `[1, 4, 6, 8]`
- Parte 2:
```python
G.remove_edge(1, 4)
G.remove_edge(4, 6)
G.remove_edge(6, 8)
```
- Parte 3: `[1, 3, 5, 8]`

### Pregunta 3

Para el sistema de red de carreteras, que se muestra en la Figura 3 determine el flujo mximo en vehculos por hora. Ahora, la comisin de carreteras considera aadir una seccin de carreteras 3-4 para permitir el flujo de 100 vehculos por hora o, a un costo adicional, un flujo de 300 vehculos por hora, por un costo mayor. Puntualizando:
1. Determine cul es el flujo mximo mximo de la red
2. Cul es su recomendacin para el arco 3-4 de la red?



**Figura 3. Grafo del Problema 3**

Fuente: Elaboracin Propia


````python
# Parte 1
G = nx.DiGraph()

G.add_edge(1, 2, capacity=4)
G.add_edge(1, 3, capacity=3)
G.add_edge(1, 5, capacity=2)

G.add_edge(2, 3, capacity=1)
G.add_edge(2, 4, capacity=2)

G.add_edge(3, 2, capacity=1)
G.add_edge(3, 5, capacity=2)
G.add_edge(3, 6, capacity=2)

G.add_edge(4, 2, capacity=2)
G.add_edge(4, 6, capacity=5)

G.add_edge(5, 3, capacity=2)
G.add_edge(5, 6, capacity=2)

flow_value, flow_dict = nx.maximum_flow(G, 1, 6)
print(flow_value)
````

````python
G.add_edge(3, 4, capacity=1)
flow_value, flow_dict = nx.maximum_flow(G, 1, 6)
print(f'100 veh/hr {flow_value}')

G.add_edge(3, 4, capacity=3)
flow_value, flow_dict = nx.maximum_flow(G, 1, 6)
print(f'300 veh/hr {flow_value}')
````

Respuestas:

1. 600 vehculos por hora
2. Al agregar, la opcin de 300 veh/hr se observa que el flujo mximo es 800, un incremento de 200 sobre el valor original, la recomendacin de hacer este tramo con esa capacidad se reduce a dos observaciones, la primera la estimacin del crecimiento del parque vehicular de la ciudad o el costo. La opnion profesional indica que al aun quedar un flujo residual de 100 unidades debera construirse el de 300 para reducir el posible congestionamiento.
