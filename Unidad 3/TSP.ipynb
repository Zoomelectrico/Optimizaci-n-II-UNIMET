## Problema del Agente Viajero (TSP)

Según Taha (2017) el problema del agente viajero (TSP) se encarga de encontrar el camino más corto para realizar un ruta cerrada (se llega a la ciudad de partida) de $n$ ciudades, en donde se pasa por cada ciudad exactamente una vez. El TSP cuenta con dos piezas de información clave:

1. El número de ciudades, denotado por $n$
2. Las distancia $d_{ij}$ entre las ciudades $i$ y $j$, se considera que $d_{ij} = \infty$ si las ciudades $i$ y $j$ no están conectadas.

El máximo número de rutas posibles para esta situación viene dado por la expresión $(n-1)!$

### Ejemplo 



```python
%matplotlib inline 
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Utilizando NetworkX
G = nx.Graph()

G.add_edge(1, 2, weight=10)
G.add_edge(1, 3, weight=15)
G.add_edge(1, 4, weight=20)

G.add_edge(2, 3, weight=35)
G.add_edge(2, 4, weight=25)

G.add_edge(3, 4, weight=30)

nx.draw(G, with_labels = True)
```

Luego haciendo uso del método [`nx.approximation.traveling_salesman_problem`](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.approximation.traveling_salesman.traveling_salesman_problem.html) vamos a resolver el problema. 


```python
paths = nx.approximation.traveling_salesman_problem(G)
print(paths)
```

🎉🎉🎉 Listo ya sabemos resolver el problema de la ruta más corta
