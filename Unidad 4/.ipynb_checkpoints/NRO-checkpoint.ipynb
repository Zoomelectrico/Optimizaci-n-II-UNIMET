## Algoritmos No Restringidos

### Búsqueda Directa

Los métodos de búsqueda directa se aplican sobre todo a funciones de una sola variable estrictamente [unimodales](https://en.wikipedia.org/wiki/Unimodality#Unimodal_function). La idea de los métodos de búsqueda directa es identificar el **intervalo de incertidumbre** que se sabe incluye el punto de solución óptima. El procedimiento localiza el óptimo estrechando de manera interactiva el intervalo de incertidumbre a un nivel de exactitud deseada.

Se presentan dos algoritmos de búsqueda estrechamente relacionados; el *dicótomo* y el de *la sección dorada*. Ambos buscan la maximización de una función unimodal $f(x)$ a lo largo del intervalo $a \le x \le b$ que incluye el punto óptimo $x^*$. Los dos métodos se inician con el intervalo inicial de incertidumbre $I_0 = \left(a,b \right)$

**Paso general $i$** 

Sea $I_{i-1} = \left(X_L, X_R  \right)$ el intervalo actual de incertidumbre. La Tabla 1 muestra cómo se determinan $x_1$ y $x_2$:

La selección de $x_1$ y $x_2$ garantiza que $X_L < x_1 < x_2 < X_R$.

El siguiente intervalo de incertidumbre, $I_i$, se determina de la siguiente manera:

1. $f(x_1) > f(x_2)$, entonces $x_L < x^* < x_2$. Sea $x_R = x_2$ y $I_i = \left(x_L,x_2 \right)$
2. $f(x_1) < f(x_2)$, entonces $x_1 < x^* < x_R$. Sea $x_L = x_1$ y $I_i = \left(x_1,x_R \right)$
3. $f(x_1) = f(x_2)$, entonces $x_1 < x^* < x_2$. Sea $x_L = x_1 \: \text{y} \:x_R = x_2$ y $I_i = \left(x_1,x_2 \right)$

La manera de determinar $x_1$ y $x_2$ garantiza que $I_{i+1} < I_i$. El algoritmo termina en la iteración $k$ si $I_k \le \Delta$, donde $\Delta$ es un nivel de exactitud especificado por el usuario.

#### Ejemplo

$$ \max{f(x)} = \begin{array}{lcc}3x &si& 0\le x \le 2\\ \frac{1}{3}\left(-x + 20\right) & si & 2 < x \le 3\end{array}$$

```python
def f(x):
    if (x <= 2):
        return 3*x
    else:
        return (1/3)*(-x+20)
    
def diff(a, b, tol):
    return (b-a) <= tol
    
def dichotomous(func, xL, xR, delta=1e-10, tolerance=1e-1, max_iters=1500):
    if (xL > xR):
        print("Ha ocurrido un error xL no puede ser mayor que xR")
        return None
    XL = xL
    XR = xR
    iters = 1;
    
    while not diff(XL, XR, tolerance) and iters < max_iters:
        x1 = (1/2) * (XL + XR - delta)
        x2 = (1/2) * (XL + XR + delta)
        if (func(x1) > func(x2)):
            XR = x2
        elif (func(x1) < func(x2)):
            XL = x1
        else:
            XL = x1
            XR = x2
        iters += 1

        
    return XL, XR, iters

print(dichotomous(f, 0, 3))    
```

```python
def f(x):
    if (x <= 2):
        return 3*x
    else:
        return (1/3)*(-x+20)
    
def diff(a, b, tool):
    return (b-a) <= tool

def golden_section(func, xL, xR, delta=0.1, tolerance=1e-1, max_iters=1500):
    if (xL > xR):
        print("Ha ocurrido un error xL no puede ser mayor que xR")
        return None
    XL = xL
    XR = xR
    iters = 1;
    
    while not diff(XL, XR, tolerance) and iters < max_iters:
        x1 = XR - (((5)**(1/2) - 1) / 2) * (XR - XL)
        x2 = XL + (((5)**(1/2) - 1) / 2) * (XR - XL)
        if (func(x1) > func(x2)):
            XR = x2
        elif (func(x1) < func(x2)):
            XL = x1
        else:
            XL = x1
            XR = x2
        iters += 1

        
    return XL, XR, iters

print(golden_section(f, 0, 3))
```

### Gradiente

El método del gradiente funciona para optimizar dos veces funciones continuamente diferenciables, también es. llamado método del *ascenso más pronunciado* (o de mayor pendiente). La idea es generar puntos sucesivos en la dirección del gradiente de la función. La terminación del método de gradiente se da en el punto donde el vector gradiente se vuelve nulo. Ésta es la única condición necesaria para la optimalidad.

Suponga que $f(\mathbf{x})$ se maximiza. Sea $X_0$ el punto inicial desde donde se inicia el procedimiento, defina $\nabla{f(\mathbf{x}_k)}$ como el gradiente de $f$ en el punto $\mathbf{x}_k$. La idea es determinar una ruta particular $p$ a lo largo de la cual $\frac{\partial{f}}{\partial{p}}$ se maximice en un punto dado. Este resultado se logra si se seleccionan los puntos sucesivos $\mathbf{X}_k$ y $\mathbf{X}_{k+1}$ de modo que:

$$\mathbf{x}_{k+1} = \mathbf{x}_{k} + r_k\nabla{f(\mathbf{x}_k)}$$

En donde $r_k$ es el tamaño del paso óptimo en $\mathbf{x}_{k}$. El tamaño del paso se determina de modo que el siguiente punto $\mathbf{x}_{k+1}$ conduzca al mejoramiento máximo de $f$. Esto equivale a determinar $r = r_k$ que maximiza la función

$$h(r) = f\left[\mathbf{X}_k + r\nabla{f(\mathbf{X}_k)}\right]$$

Debido a que $h(r)$ es una función de una sola variable, se puede utilizar el método de búsqueda (anterior) para determinar el óptimo, siempre que $h(r)$ sea unimodal.

El procedimiento propuesto termina cuando dos puntos de prueba sucesivos $\mathbf{X}_k$ y $\mathbf{X}_{k+1}$ son aproximadamente iguales. Esto equivale a tener $r_k\nabla{f(\mathbf{X}_k)}\approx 0$

#### Ejemplo

$$
    f(x_1, x_2) = 2x_1^2+2x_2^2-2x_1x_2 + 10\\[8pt]
    \nabla{f(x_1,x_2)} = \left(4x_1-2x_2, \: 4x_2-2x_1\right)
$$

```python
%matplotlib inline
import numpy as np
import numpy.linalg as la
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as pt
from mpl_toolkits.mplot3d import axes3d

def f(x):
    return 2*x[0]**2 + 2*x[1]**2 - 2*x[0]*x[1] + 10

def df(x):
    return np.array([4*x[0]-2*x[1], 4*x[1]-2*x[0]])


fig = pt.figure()
ax = fig.gca(projection="3d")

xmesh, ymesh = np.mgrid[-100:100,-100:100]
fmesh = f(np.array([xmesh, ymesh]))
ax.plot_surface(xmesh, ymesh, fmesh)
```

```python
pt.axis("equal")
pt.contour(xmesh, ymesh, fmesh)
```

```python
guesses = [np.array([100, 100])]
```

```python
guesses
```

```python
iters = 0

while iters < 10:
    x = guesses[-1]
    s = -df(x)
    def f1d(alpha):
        return f(x + alpha*s)
    alpha_opt = minimize_scalar(f1d, bracket=(0, 1)).x
    next_guess = x + alpha_opt * s
    guesses.append(next_guess)
    iters += 1
```

```python
pt.axis("equal")
pt.contour(xmesh, ymesh, fmesh, 20)
it_array = np.array(guesses)
pt.plot(it_array.T[0], it_array.T[1], "x-")
```
