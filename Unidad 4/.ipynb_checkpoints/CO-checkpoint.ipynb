## Problemas no restringidos

### Ejemplo

Consiedere la función $f(x,y)$ determine:

1. Gráfique la función $f$
2. Los puntos estacionarios de la función
3. Si es un minímo o un máximo

$$ 
f(x,y) = x^3 + 2y^2 + 3xy^2
$$

```python
# Parte 1y
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creamos la figura y los ejes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definimos la función 
def f(x,y):
    return x**3 + 2*y**2 + 3*x*y**2

# Definimos valores x, y para calcular z
x = y = np.arange(-5.0, 5.0, 0.05)
X, Y = np.meshgrid(x, y)
z = np.array(f(np.ravel(X), np.ravel(Y)))
Z = z.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# Parte 2
from __future__ import division
from sympy import *

x, y = symbols('x y')

f = x**3 + 2*y**2 + 3*x*y**2

fx = diff(f, x)
fy = diff(f, y)
print('Derivadas Parciales de 1er orden:')
print(f'fx: {fx}')
print(f'fy: {fy}')

print(solve(fx))

# Parte 3

pprint(hessian(f, (x,y)))
```

### Método de Newton-Raphson

#### Ejemplo

$$f(x) = (3x-2)^2(2x-3)^2$$

```python
from scipy.optimize import newton

def f(x):
    return (3*x - 2)**2 * (2*x-3)**2

y = symbols('y')

print(diff((3*y - 2)**2 * (2*y-3)**2, y))

x = np.arange(-10, 10, 0.05)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, f(x))

root = newton(f, 10)
ax.plot(root, f(root), '.')
root = newton(f, 0.5)
ax.plot(root, f(root), '.')
root = newton(f, 1)
ax.plot(root, f(root), '.')

plt.show()
```
